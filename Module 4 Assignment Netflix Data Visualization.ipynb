## Create visualizations to represent the Data on Netflix

```python
#Most watched Genres on Netflix

colors= ['#b6c2d6','#49576e']
plt.figure(figsize=(15,10))
df.type.value_counts().plot(kind='pie', colors=colors, autopct='%.2f%%')
plt.title('Most Watched Genres On Netflix')
plt.legend()
plt.show()
```

```python
#Ploting top Ratings On Netflix

plt.figure(figsize=(15,10))
sns.countplot(df.rating,color="#585c28")
plt.title('Ratings distribution On Netflix ')
plt.show()
```

```python
#Rating By Type of Movies On Netflix

plt.figure(figsize=(15,10))
sns.countplot(x='rating', hue="type", data=df)
plt.title("Rating Distribution On Netflix Based On movies Type")
plt.show()
```

```python
#Top 10 Countries released Shows On Netflix
plt.figure(figsize=(15,10))
df.country.value_counts().head(10).plot(kind='bar', color="#329ba8")
plt.title("Top 10 Movies Released Shows on Netflix")
plt.xlabel("Countries")
plt.ylabel("Counts")
plt.show()
```

```python
# How many movies are released in each year?
# Top 10 years in which most movies are released
plt.figure(figsize=(15,10))
df.release_year.value_counts().head(10).plot(kind='bar',color='#828c7a')
plt.title("Top 10 years in which most movies are released")
plt.xlabel("Release year")
plt.ylabel("Counts")
plt.legend()
plt.show()
```

```python
# In which month most of the shows release?
plt.figure(figsize=(15,10))
df['date_added'].value_counts().head(10).plot(kind='bar')
plt.title("Months with Highest Movies Released")
plt.legend()
plt.xlabel("Months")
plt.show()
```

```python
plt.figure(figsize=(15,10))
df.release_year.value_counts().head(10).plot(kind='pie',color='#998e98',autopct='%.2f%%')
plt.title("Top 10 years in which most TV Shows are released")
#plt.xlabel("Release year")
#plt.ylabel("Counts")
plt.legend()
plt.show()
```
